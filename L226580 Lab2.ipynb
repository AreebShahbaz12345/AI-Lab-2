```python
import numpy as np
import matplotlib.pyplot as plt

# Read the genome sequence from a file
with open('genome_sequence.txt', 'r') as file:
    genome_sequence = file.read().strip()

# Convert the sequence to a list and find its length
genome_list = list(genome_sequence)
genome_length = len(genome_list)

# Parametric equations for a helix
t = np.linspace(0, 4 * np.pi, genome_length)  # 4*pi gives about 2 turns
x = np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5, genome_length)  # z increases linearly

# Combine the coordinates into a (genome_length x 3) array
coordinates = np.column_stack((x, y, z))

# Assign colors to each molecule (example: A=red, T=green, C=blue, G=yellow)
color_map = {'A': 'red', 'T': 'green', 'C': 'blue', 'G': 'yellow'}
colors = [color_map.get(base, 'black') for base in genome_list]

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c=colors)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Helix Structure of Genome Sequence')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Load an image from the internet and convert it to a NumPy array
image_url = 'Sample.jpg'  # Replace with your image URL
image = Image.open(image_url)
img_array = np.array(image)

# Plot the original image
plt.figure(figsize=(10, 5))
plt.subplot(2, 2, 1)
plt.imshow(img_array)
plt.title('Original Image')

# Rotate the image by 90 degrees
rotated_img = np.rot90(img_array)
plt.subplot(2, 2, 2)
plt.imshow(rotated_img)
plt.title('Rotated 90 Degrees')

# Flip the image horizontally
flipped_img = np.fliplr(img_array)
plt.subplot(2, 2, 3)
plt.imshow(flipped_img)
plt.title('Flipped Horizontally')

# Convert the image to grayscale
gray_img = np.dot(img_array[..., :3], [0.299, 0.587, 0.114])
plt.subplot(2, 2, 4)
plt.imshow(gray_img, cmap='gray')
plt.title('Grayscale Image')

# Show the plots
plt.tight_layout()
plt.show()
```
